```python
import pandas as pd
import numpy as np
```

```python
excelfile=pd.ExcelFile('TakenMind.xlsx')
```

```python
df=excelfile.parse('Employees who have left')
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df).savefig('image1.png')
```

```python
import matplotlib as mp
import matplotlib.pyplot as plt
grid = sns.FacetGrid(df, row="promotion_last_5years", col="time_spend_company", margin_titles=True)
grid.map(plt.hist, "Emp ID", bins=np.linspace(0, 40, 15)).savefig('image2.png')
```

```python
with sns.axes_style('white'):
    sns.jointplot("Emp ID", "satisfaction_level", data=df, kind='hex').savefig('image3.png')
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("salary", data=df, aspect=2,
                       kind="count", color='steelblue').savefig('image4.png')
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("dept", data=df, aspect=2,
                       kind="count", color='steelblue').savefig('image5.png')
```

```python
# calculate the correlation matrix
corr = df.corr()

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
plt.savefig('image8.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="average_montly_hours", y="satisfaction_level",
                hue="dept",
                sizes=(1, 8), linewidth=0,
                data=df)
plt.savefig('image99.png')
```

```python
g = sns.catplot(x="dept", y="average_montly_hours", hue="salary", data=df,
                height=15, kind="bar", palette="muted")
g.despine(left=True)
plt.savefig('image101.png')
```

```python
sns.violinplot(x="number_project", y="satisfaction_level",
               data=df)
plt.savefig('imagica.png')
```

```python
f, ax = plt.subplots(figsize=(12, 8))
sns.boxplot(x="dept", y="number_project",
               data=df)
plt.savefig('imagicalul.png')
```
